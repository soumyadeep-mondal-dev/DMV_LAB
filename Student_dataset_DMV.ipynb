```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Load Dataset
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
print("Dataset Preview:")
print(df.head())
```

```python
# Handle Missing Values
print("\nMissing Values Before:")
print(df.isnull().sum())
```

```python
# Fill numeric with mean
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Fill categorical with mode
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing Values After:")
print(df.isnull().sum())
```

```python
# Outlier Detection (Boxplot)
plt.figure()
df.select_dtypes(include='number').plot(kind='box')
plt.title("Outlier Detection (Boxplot)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Get numeric columns
num_cols = df.select_dtypes(include='number').columns
```

```python
# Bar Chart
plt.figure()
df.iloc[:,0].value_counts().plot(kind='bar')
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# Scatter Plot
if len(num_cols) >= 2:
    plt.figure()
    plt.scatter(df[num_cols[0]], df[num_cols[1]])
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.title("Scatter Plot")
    plt.show()
```

```python
# stair chart
plt.figure(figsize=(12,6))
plt.step(df["student_id"],df["risk_score"],where='mid')
plt.xlabel("Student ID")
plt.ylabel("Rish Score")
plt.title("stair chart")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Line Chart
plt.figure()
df[num_cols[0]].plot()
plt.title("Line Chart")
plt.xlabel("Index")
plt.ylabel(num_cols[0])
plt.show()
```

```python
# Pie Chart
plt.figure()
df.iloc[:,0].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart")
plt.ylabel("")
plt.show()
```

```python
# Histogram
plt.figure()
df[num_cols[0]].hist(bins=10)
plt.title("Histogram")
plt.xlabel(num_cols[0])
plt.ylabel("Frequency")
plt.show()
```

```python
# Save Cleaned Dataset
df.to_csv("cleaned_data.csv", index=False)

print("\n✅ All tasks completed successfully!")
```
